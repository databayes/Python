```python
df = spark.read.format("csv").option("header","true").load("abfss://1fe39462-9846-4fe4-93dd-46978e9def3a@onelake.dfs.fabric.microsoft.com/a1121598-372c-4ac8-bc11-78500c7274bc/Files/HR_file.csv")
# df now is a Spark DataFrame containing CSV data from "abfss://1fe39462-9846-4fe4-93dd-46978e9def3a@onelake.dfs.fabric.microsoft.com/a1121598-372c-4ac8-bc11-78500c7274bc/Files/HR_file.csv".
display(df)
```

```python
# Code generated by Data Wrangler for pandas sample

def clean_data(pandas_df):
    # Rename column 'average_montly_hours' to 'average_monthly_hours'
    pandas_df = pandas_df.rename(columns={'average_montly_hours': 'average_monthly_hours'})
    # Drop duplicate rows across all columns
    pandas_df = pandas_df.drop_duplicates()
    # Replace missing values with "0" in columns: 'satisfaction_level', 'last_evaluation', 'time_spend_company'
    pandas_df = pandas_df.fillna({'satisfaction_level': "0", 'last_evaluation': "0", 'time_spend_company': "0"})
    # Capitalize the first character in columns: 'Departments', 'salary'
    pandas_df['Departments'] = pandas_df['Departments'].str.title()
    pandas_df['salary'] = pandas_df['salary'].str.title()
    # Performed 2 aggregations grouped on columns: 'Departments', 'salary'
    pandas_df = pandas_df.groupby(['Departments', 'salary']).agg(promotion_last_5years_count=('promotion_last_5years', 'count'), number_project_mode=('number_project', lambda s: s.value_counts().index[0])).reset_index()
    return pandas_df

# Loaded variable 'df' from kernel state
pandas_df = df.limit(5000).toPandas()

pandas_df_clean = clean_data(pandas_df.copy())
pandas_df_clean.head(20)
```
